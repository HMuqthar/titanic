```python
train_len=len(train)
#combine two dataframe
df=pd.concat((train,test),axis=0)
df=df.reset_index(drop=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df=df.drop(columns=['Cabin'],axis=0) 
```

```python
df.head()
```

```python
df['Age'].mean()
```

```python
#fill missing null values with mean for numerical column

df["Age"]=df["Age"].fillna(df["Age"].mean())
df["Fare"]=df["Fare"].fillna(df["Fare"].mean())
df
```

```python
#fill missing values using mode for categorical coloumn
df['Embarked'].mode()[0]
```

```python
df['Embarked']=df['Embarked'].fillna(df['Embarked'].mode()[0])
df
```

```python
df.info()
```

```python
#log transformation for uniform data distribution
sns.distplot(df['Fare'])
```

```python
df['Fare']=np.log(df['Fare']+1)
```

```python
sns.distplot(df['Fare'])
```

```python
corr=df.corr()
plt.figure(figsize=(15,10))
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
#drop the unneccessary coloumn
df=df.drop(columns=['Name','Ticket'],axis=0)
```

```python
df.head()
```

```python
#convert sex and embark to number using mapping
#label encoding
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
cols=['Sex','Embarked']
for col in cols:
    df[col]=le.fit_transform(df[col])
df.head()
```

```python
df.tail()
```

```python
#split into training and testing
train=df.iloc[:train_len,:]
test=df.iloc[train_len:]
train.head()
```

```python
test.head()
```

```python
#input split
x=train.drop(columns=['PassengerId','Survived'],axis=1)
y=train['Survived']
x.head()
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
#classify column
def classify(model,x,y):
        x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.25,random_state=42)
        model.fit(x_train, y_train)
        print('Accur
              acy:',model.score(x_test,y_test))
        score=cross_val_score(model,x,y,cv=5)
        print('CV Score:',np.mean(score))
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
classify(model,x,y)
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier()
classify(model,x,y)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
classify(model,x,y)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
classify(model,x,y)
```
